```python
import pandas as pd
path = "/media/zunayed/HDD_code2/book-recommender/data/books.csv"
books = pd.read_csv(path)
books.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14, 6))
ax = plt.axes()

sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Rows with Missing Values (Highlighted)")
plt.title("Missing Data Heatmap")
plt.show()
```

```python
import numpy as np
```

```python
books["missing_description"] = np.where(books["description"].isna(), 1, 0)
```

```python
books["age_of_book"] = np.where(books["published_year"].isna(), 1, 0)
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]
correlation_matrix = books[columns_of_interest].corr(method="spearman")
sns.set_theme(style="white")
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "spearman"})

heatmap.set_title("correlation Heatmap")
plt.show()
```

```python
books[
    (books["description"].isna()) |
    (books["missing_description"].isna()) |
    (books["average_rating"].isna()) |
    (books["published_year"].isna())
]
```

```python
book_missing = books[
    ~(books["description"].isna()) &
    ~(books["num_pages"].isna()) &
    ~(books["average_rating"].isna()) &
    ~(books["published_year"].isna())
]
book_missing
```

```python
book_missing['categories'].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
book_missing
```

```python
book_missing.loc[book_missing["words_in_description"].between(1, 4), "description"]
```

```python
 book_missing.loc[book_missing["words_in_description"].between(5, 14), "description"]
```

```python
book_missing.loc[book_missing["words_in_description"].between(25, 35
                                                        ), "description"]
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] > 25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = np.where(
    book_missing_25_words["subtitle"].isna(),
    book_missing_25_words["title"],
    book_missing_25_words[["title", "subtitle"]].fillna("").astype(str).agg(": ".join, axis=1)
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words.drop(
        ["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1
    )
    .to_csv("books_cleadned.csv", index=False)
)
```
